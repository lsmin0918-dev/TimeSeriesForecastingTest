````python
# ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Îî•Îü¨ÎãùÏö©)
sequence_length = 30

def create_sequences(X, y, seq_len=30):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_len):
        X_seq.append(X[i:i+seq_len])
        y_seq.append(y[i+seq_len])
    return np.array(X_seq), np.array(y_seq)

X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train.values, sequence_length)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, y_val.values, sequence_length)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test.values, sequence_length)

print(f"ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞ shape: {X_train_seq.shape}")
print(f"  - ÏÉòÌîå Ïàò: {X_train_seq.shape[0]}")
print(f"  - ÏãúÌÄÄÏä§ Í∏∏Ïù¥: {X_train_seq.shape[1]}")
print(f"  - ÌäπÏÑ± Ïàò: {X_train_seq.shape[2]}")
````

## 2. ÏòàÏ†ú: LSTM Î™®Îç∏ + ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à Ï†ÑÎûµ

Ïù¥ ÏòàÏ†úÎäî 2Í∞ÄÏßÄ ÌïµÏã¨ ÏöîÏÜåÎ°ú Íµ¨ÏÑ±Îê©ÎãàÎã§:

1. **ÏòàÏ∏° Î™®Îç∏**: Í∞ÄÍ≤© ÏÉÅÏäπ/ÌïòÎùΩÏùÑ ÏòàÏ∏°ÌïòÍ≥† ÌôïÎ•† Ï†úÍ≥µ
2. **Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ**: ÏòàÏ∏° ÌôïÎ•†Ïóê Îî∞Îùº Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à

### Ï†ÑÎûµ ÏòàÏãú
- ÏÉÅÏäπ ÌôïÎ•† 90% ‚Üí ÏûêÎ≥∏Ïùò 90% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 60% ‚Üí ÏûêÎ≥∏Ïùò 60% Ìà¨Ïûê
- ÏÉÅÏäπ ÌôïÎ•† 40% ‚Üí Ìà¨ÏûêÌïòÏßÄ ÏïäÏùå (ÌòÑÍ∏à Î≥¥Ïú†)

````python
# ÏòàÏ†ú Î™®Îç∏: LSTM Í∏∞Î∞ò 2-class Î∂ÑÎ•òÍ∏∞
class ExampleLSTMModel(nn.Module):
    """
    ÏòàÏ†ú: LSTM Í∏∞Î∞ò Í∞ÄÍ≤© Î∞©Ìñ• ÏòàÏ∏° Î™®Îç∏
    Ï∂úÎ†•: ÏÉÅÏäπ ÌôïÎ•† (0~1)
    """
    def __init__(self, input_size, hidden_size=64, dropout=0.2):
        super(ExampleLSTMModel, self).__init__()
        
        self.lstm1 = nn.LSTM(input_size, hidden_size, batch_first=True, num_layers=1)
        self.dropout1 = nn.Dropout(dropout)
        self.bn1 = nn.BatchNorm1d(hidden_size)
        
        self.lstm2 = nn.LSTM(hidden_size, hidden_size//2, batch_first=True, num_layers=1)
        self.dropout2 = nn.Dropout(dropout)
        self.bn2 = nn.BatchNorm1d(hidden_size//2)
        
        self.fc1 = nn.Linear(hidden_size//2, 16)
        self.relu = nn.ReLU()
        self.dropout3 = nn.Dropout(dropout)
        
        self.fc2 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        lstm_out, _ = self.lstm1(x)
        lstm_out = self.dropout1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        lstm_out = self.bn1(lstm_out)
        lstm_out = lstm_out.permute(0, 2, 1)
        
        lstm_out, _ = self.lstm2(lstm_out)
        lstm_out = self.dropout2(lstm_out[:, -1, :])
        lstm_out = self.bn2(lstm_out)
        
        out = self.fc1(lstm_out)
        out = self.relu(out)
        out = self.dropout3(out)
        
        out = self.fc2(out)
        out = self.sigmoid(out)
        
        return out

# ÏòàÏ†ú Î™®Îç∏ ÏÉùÏÑ±
example_model = ExampleLSTMModel(
    input_size=X_train_seq.shape[2],
    hidden_size=64,
    dropout=0.2
).to(device)

print("ÏòàÏ†ú Î™®Îç∏ Íµ¨Ï°∞:")
print(example_model)
print(f"\nTotal parameters: {sum(p.numel() for p in example_model.parameters()):,}")
````

````python
# ÌïôÏäµ Ìï®Ïàò (Ï†úÍ≥µÎê®)
def train_model(model, train_loader, val_loader, epochs=100, lr=0.001, patience=15):
    model = model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_loss = float('inf')
    patience_counter = 0
    best_model_state = None
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y.unsqueeze(1))
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            predicted = (outputs > 0.5).float()
            train_total += batch_y.size(0)
            train_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_train_loss = train_loss / len(train_loader)
        train_acc = train_correct / train_total
        
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(device), batch_y.to(device)
                outputs = model(batch_X)
                loss = criterion(outputs, batch_y.unsqueeze(1))
                
                val_loss += loss.item()
                predicted = (outputs > 0.5).float()
                val_total += batch_y.size(0)
                val_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_val_loss = val_loss / len(val_loader)
        val_acc = val_correct / val_total
        
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}')
        
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break
    
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
    
    return history

print("‚úÖ ÌïôÏäµ Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

````python
# ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ
print("ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏãúÏûë...")
example_history = train_model(
    model=example_model,
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=100,
    lr=0.001,
    patience=15
)
print("\n‚úÖ ÏòàÏ†ú Î™®Îç∏ ÌïôÏäµ ÏôÑÎ£å!")
````

````python
# ÏòàÏ∏° Ìï®Ïàò (ÌôïÎ•† Ìè¨Ìï®)
def predict_with_probability(model, data_loader):
    """
    Î™®Îç∏ ÏòàÏ∏° Î∞è ÌôïÎ•† Î∞òÌôò
    """
    model.eval()
    predictions_prob = []
    
    with torch.no_grad():
        for batch_X, _ in data_loader:
            batch_X = batch_X.to(device)
            outputs = model(batch_X)
            predictions_prob.append(outputs.cpu().numpy())
    
    predictions_prob = np.vstack(predictions_prob).flatten()
    predictions = (predictions_prob > 0.5).astype(int)
    
    return predictions_prob, predictions

# ÏòàÏ†ú Î™®Îç∏ ÏòàÏ∏°
example_prob, example_pred = predict_with_probability(example_model, test_loader)

print(f"ÏòàÏ∏° ÏôÑÎ£å!")
print(f"ÏòàÏ∏° Ïàò: {len(example_pred)}")
print(f"ÏÉÅÏäπ ÏòàÏ∏°: {np.sum(example_pred == 1)}Í∞ú")
print(f"ÌïòÎùΩ ÏòàÏ∏°: {np.sum(example_pred == 0)}Í∞ú")
print(f"\nÌèâÍ∑† ÏÉÅÏäπ ÌôïÎ•†: {example_prob.mean():.2%}")
````

````python
# ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ
def simulate_probability_trading(predictions_prob, actual_prices, dates, 
                                 initial_capital=10000, transaction_fee=0.001,
                                 threshold=0.5, position_scaling=True):
    """
    ÌôïÎ•† Í∏∞Î∞ò Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò
    
    Parameters:
    -----------
    predictions_prob : array
        ÏÉÅÏäπ ÌôïÎ•† (0~1)
    actual_prices : array
        Ïã§Ï†ú Í∞ÄÍ≤©
    dates : array
        ÎÇ†Ïßú
    initial_capital : float
        Ï¥àÍ∏∞ ÏûêÎ≥∏
    transaction_fee : float
        Í±∞Îûò ÏàòÏàòÎ£å
    threshold : float
        Îß§Ïàò Í≤∞Ï†ï ÏûÑÍ≥ÑÍ∞í (Í∏∞Î≥∏ 0.5)
    position_scaling : bool
        True: ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê ÎπÑÏú® Ï°∞Ï†à
        False: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
    """
    cash = initial_capital
    btc_holdings = 0
    portfolio_values = []
    trade_log = []
    
    for i in range(len(predictions_prob)):
        current_price = actual_prices[i]
        prob = predictions_prob[i]
        
        portfolio_value = cash + btc_holdings * current_price
        portfolio_values.append(portfolio_value)
        
        # ÎßàÏßÄÎßâ ÎÇ† Ï†ÑÎüâ Îß§ÎèÑ
        if i == len(predictions_prob) - 1:
            if btc_holdings > 0:
                sell_value = btc_holdings * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL_ALL',
                    'price': current_price,
                    'prob': prob,
                    'amount': btc_holdings,
                    'value': btc_holdings * current_price,
                    'fee': btc_holdings * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings = 0
            continue
        
        # Ìà¨Ïûê ÎπÑÏú® Í≤∞Ï†ï
        if position_scaling:
            # ÌôïÎ•† Í∏∞Î∞ò Ìè¨ÏßÄÏÖò Ï°∞Ï†à
            if prob > threshold:
                # ÌôïÎ•†Ïóê ÎπÑÎ°ÄÌïòÏó¨ Ìà¨Ïûê (Ïòà: 70% ÌôïÎ•† ‚Üí 70% Ìà¨Ïûê)
                invest_ratio = prob
            else:
                # ÏûÑÍ≥ÑÍ∞í ÎØ∏ÎßåÏù¥Î©¥ Îß§ÎèÑ ÎòêÎäî Í¥ÄÎßù
                invest_ratio = 0
        else:
            # Îã®Ïàú Ï†ÑÎûµ: ÏûÑÍ≥ÑÍ∞í Ï¥àÍ≥º Ïãú Ï†ÑÏï° Ìà¨Ïûê
            invest_ratio = 1.0 if prob > threshold else 0.0
        
        # ÌòÑÏû¨ Ìè¨ÏßÄÏÖò ÎπÑÏú®
        current_position_ratio = (btc_holdings * current_price) / portfolio_value if portfolio_value > 0 else 0
        
        # Ìè¨ÏßÄÏÖò Ï°∞Ï†ï
        target_btc_value = portfolio_value * invest_ratio
        current_btc_value = btc_holdings * current_price
        
        if target_btc_value > current_btc_value:  # Îß§Ïàò ÌïÑÏöî
            buy_cash = target_btc_value - current_btc_value
            if buy_cash > cash:
                buy_cash = cash
            if buy_cash > 0:
                buy_amount = (buy_cash * (1 - transaction_fee)) / current_price
                btc_holdings += buy_amount
                trade_log.append({
                    'date': dates[i],
                    'action': 'BUY',
                    'price': current_price,
                    'prob': prob,
                    'amount': buy_amount,
                    'value': buy_cash,
                    'fee': buy_cash * transaction_fee
                })
                cash -= buy_cash
        elif target_btc_value < current_btc_value:  # Îß§ÎèÑ ÌïÑÏöî
            sell_btc = (current_btc_value - target_btc_value) / current_price
            if sell_btc > btc_holdings:
                sell_btc = btc_holdings
            if sell_btc > 0:
                sell_value = sell_btc * current_price * (1 - transaction_fee)
                trade_log.append({
                    'date': dates[i],
                    'action': 'SELL',
                    'price': current_price,
                    'prob': prob,
                    'amount': sell_btc,
                    'value': sell_btc * current_price,
                    'fee': sell_btc * current_price * transaction_fee
                })
                cash += sell_value
                btc_holdings -= sell_btc
    
    final_value = portfolio_values[-1]
    total_return = (final_value - initial_capital) / initial_capital * 100
    
    total_trade_volume = sum(trade['value'] for trade in trade_log)
    total_fees_paid = sum(trade['fee'] for trade in trade_log)
    
    return {
        'initial_capital': initial_capital,
        'final_value': final_value,
        'total_return': total_return,
        'portfolio_values': portfolio_values,
        'trade_log': trade_log,
        'num_trades': len(trade_log),
        'total_trade_volume': total_trade_volume,
        'total_fees_paid': total_fees_paid
    }

print("‚úÖ Ìä∏Î†àÏù¥Îî© ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìï®Ïàò Ï§ÄÎπÑ ÏôÑÎ£å!")
````

```
     3) MACD Í≥®Îì†ÌÅ¨Î°úÏä§ Ïãú Ìà¨Ïûê ÎπÑÏ§ë ÌôïÎåÄ
   - **Ï≤≠ÏÇ∞/ÎπÑÏ§ëÏ∂ïÏÜå**:
     1) Î™®Îç∏ ÌôïÎ•† < 0.4
     2) RSI > 70 (Í≥ºÎß§Ïàò)
     3) Ï¥àÍ≥†Î≥ÄÎèôÏÑ±(ÏÉÅÏúÑ 10%) Î∞úÏÉù Ïãú ÎπÑÏ§ë Ï∂ïÏÜå

4. ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞:
   - BUY_Threshold: 0.50
   - RSI_Oversold: 30
   - Volatility_Filter: Top 10%

5. ÏòàÏ†úÏôÄÏùò Ï∞®Î≥ÑÏ†ê:
   - ÏòàÏ†ú Î™®Îç∏ÏùÄ 'ÌôïÎ•†'Îßå Î≥¥ÏßÄÎßå, Ïù¥ Ï†ÑÎûµÏùÄ ÏãúÏû•Ïùò Í≥ºÏó¥/Ïπ®Ï≤¥(RSI)ÏôÄ Ï∂îÏÑ∏(MACD)Î•º Ìï®Íªò Í≥†Î†§Ìï®.
   - ÌäπÌûà ÌïòÎùΩÏû•ÏóêÏÑúÎèÑ Í∏∞Ïà†Ï†Å Î∞òÎì±(Dead Cat Bounce)ÏùÑ Ïû°ÏïÑÎÇ¥Í∏∞ ÏúÑÌï¥ RSI Í≥ºÎß§ÎèÑ Ïã†Ìò∏Î•º Ï†ÅÍ∑π ÌôúÏö©Ìï®.
```

````python
# MyTradingModel: RobustGRU Íµ¨ÌòÑ (Improved over ExampleLSTM)
class MyTradingModel(nn.Module):
    """
    RobustGRU: 2-layer GRU with Dropout & BatchNorm
    """
    def __init__(self, input_size, hidden_size=128, dropout=0.3):
        super(MyTradingModel, self).__init__()
        
        # Ï≤´ Î≤àÏß∏ GRU Î†àÏù¥Ïñ¥
        self.gru1 = nn.GRU(input_size, hidden_size, batch_first=True, num_layers=1, bidirectional=False)
        self.dropout1 = nn.Dropout(dropout)
        self.bn1 = nn.BatchNorm1d(hidden_size)
        
        # Îëê Î≤àÏß∏ GRU Î†àÏù¥Ïñ¥ (ÌÅ¨Í∏∞ Ï∂ïÏÜå)
        self.gru2 = nn.GRU(hidden_size, hidden_size//2, batch_first=True, num_layers=1)
        self.dropout2 = nn.Dropout(dropout)
        self.bn2 = nn.BatchNorm1d(hidden_size//2)
        
        # ÏôÑÏ†Ñ Ïó∞Í≤∞Ï∏µ (FC)
        self.fc1 = nn.Linear(hidden_size//2, 32)
        self.relu = nn.ReLU()
        self.dropout3 = nn.Dropout(dropout)
        
        self.fc2 = nn.Linear(32, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        # x shape: (batch, seq, feature)
        out, _ = self.gru1(x)
        out = self.dropout1(out)
        
        # BatchNormÏùÑ ÏúÑÌïú Ï∞®Ïõê Î≥ÄÍ≤Ω (batch, hidden, seq)
        out = out.permute(0, 2, 1)
        out = self.bn1(out)
        out = out.permute(0, 2, 1)
        
        out, _ = self.gru2(out)
        
        # ÎßàÏßÄÎßâ ÌÉÄÏûÑÏä§ÌÖùÎßå ÏÇ¨Ïö©
        out = out[:, -1, :] 
        
        out = self.dropout2(out)
        out = self.bn2(out)
        
        out = self.fc1(out)
        out = self.relu(out)
        out = self.dropout3(out)
        
        out = self.fc2(out)
        out = self.sigmoid(out)
        return out

# Î™®Îç∏ ÏÉùÏÑ± (GPU ÏÇ¨Ïö©)
my_model = MyTradingModel(
    input_size=X_train_seq.shape[2],
    hidden_size=128,
    dropout=0.3
).to(device)

print("‚úÖ ÎÇòÏùò Î™®Îç∏(RobustGRU) ÏÉùÏÑ± ÏôÑÎ£å!")
print(my_model)
print(f"\nTotal parameters: {sum(p.numel() for p in my_model.parameters()):,}")
````

````python
# Ïª§Ïä§ÌÖÄ ÌïôÏäµ Ìï®Ïàò (Weight Decay Ï†ÅÏö©ÌïòÏó¨ Í≥ºÏ†ÅÌï© Î∞©ÏßÄ)
def train_model_custom(model, train_loader, val_loader, epochs=100, lr=0.001, patience=15):
    model = model.to(device)
    criterion = nn.BCELoss()
    # L2 Í∑úÏ†ú(Weight Decay) Ï∂îÍ∞Ä
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5) 
    
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_loss = float('inf')
    patience_counter = 0
    best_model_state = None
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y.unsqueeze(1))
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            predicted = (outputs > 0.5).float()
            train_total += batch_y.size(0)
            train_correct += (predicted.squeeze() == batch_y).sum().item()
            
        avg_train_loss = train_loss / len(train_loader)
        train_acc = train_correct / train_total
        
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(device), batch_y.to(device)
                outputs = model(batch_X)
                loss = criterion(outputs, batch_y.unsqueeze(1))
                val_loss += loss.item()
                
                predicted = (outputs > 0.5).float()
                val_total += batch_y.size(0)
                val_correct += (predicted.squeeze() == batch_y).sum().item()
        
        avg_val_loss = val_loss / len(val_loader)
        val_acc = val_correct / val_total
        
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}')
            
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break
            
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
    return history

print("ÎÇòÏùò Î™®Îç∏ ÌïôÏäµ ÏãúÏûë...")
my_history = train_model_custom(
    model=my_model,
    train_loader=train_loader,
    val_loader=val_loader,
    epochs=100,
    lr=0.0005,
    patience=20
)
print("\n‚úÖ Î™®Îç∏ ÌïôÏäµ ÏôÑÎ£å!")
````

````python
# ÏòàÏ∏° Ìï®Ïàò
def predict_with_prob(model, loader):
    model.eval()
    probs = []
    with torch.no_grad():
        for batch_X, _ in loader:
            batch_X = batch_X.to(device)
            out = model(batch_X)
            probs.append(out.cpu().numpy())
    return np.vstack(probs).flatten()

# ÌÖåÏä§Ìä∏ ÏÖã ÏòàÏ∏° ÏàòÌñâ
my_prob = predict_with_prob(my_model, test_loader)

# ÏãúÎÆ¨Î†àÏù¥ÏÖòÏö© Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
test_start_idx = len(btc_features) - len(y_test) + sequence_length
test_prices = btc_features['Close'].iloc[test_start_idx:test_start_idx+len(my_prob)].values
test_dates = btc_features.index[test_start_idx:test_start_idx+len(my_prob)]

# Î≥¥Ï°∞ ÏßÄÌëú Ï§ÄÎπÑ (RSI, MACD)
test_rsi = btc_features['RSI_14'].iloc[test_start_idx:test_start_idx+len(my_prob)].values
test_macd = btc_features['MACD'].iloc[test_start_idx:test_start_idx+len(my_prob)].values
test_macd_sig = btc_features['MACD_Signal'].iloc[test_start_idx:test_start_idx+len(my_prob)].values

# Î≥ÄÎèôÏÑ± Îç∞Ïù¥ÌÑ∞
test_volatility = btc_features['Volatility_10'].iloc[test_start_idx:test_start_idx+len(my_prob)].values
volatility_threshold = np.percentile(test_volatility, 90) # ÏÉÅÏúÑ 10%Îßå ÌîºÌï® (Í≥µÍ≤©Ï†Å)

# Í≥µÍ≤©Ï†Å ÌïòÏù¥Î∏åÎ¶¨Îìú Ï†ÑÎûµ ÏãúÎÆ¨Î†àÏù¥ÏÖò
def simulate_custom_strategy(probs, prices, dates, rsi, macd, macd_sig, volatilities, vol_thresh, initial_capital=10000):
    cash = initial_capital
    btc = 0
    tx_fee = 0.001
    history = []
    trade_log = []
    
    # ÌååÎùºÎØ∏ÌÑ∞ : Í±∞Îûò ÌôúÏÑ±ÌôîÎ•º ÏúÑÌï¥ ÏûÑÍ≥ÑÍ∞í ÎåÄÌè≠ ÎÇÆÏ∂§
    BUY_THRESH = 0.50    # 50% Ïù¥ÏÉÅÏù¥Î©¥ Îß§Ïàò Í≤ÄÌÜ†
    SELL_THRESH = 0.40   # 40% ÎØ∏ÎßåÏù¥Î©¥ Îß§ÎèÑ Í≤ÄÌÜ†
    
    for i in range(len(probs)):
        prob = probs[i]
        price = prices[i]
        vol = volatilities[i]
        r_val = rsi[i]
        m_val = macd[i]
        m_sig = macd_sig[i]
        
        portfolio_val = cash + btc * price
        
        # ÎßàÏßÄÎßâ ÎÇ† Ï†ÑÎüâ Îß§ÎèÑ
        if i == len(probs) - 1:
            if btc > 0:
                cash += btc * price * (1 - tx_fee)
                btc = 0
            history.append(cash)
            continue
            
        # 1. Í∏∞Î≥∏ ÌÉÄÍ≤ü ÎπÑÏ§ë Í≥ÑÏÇ∞ (ÌôïÎ•† Í∏∞Î∞ò)
        # (Prob - 0.4) * 2 => 0.4ÏùºÎïå 0, 0.9ÏùºÎïå 1.0
        raw_ratio = (prob - 0.4) * 2.0
        target_ratio = min(max(raw_ratio, 0.0), 1.0)
        
        # 2. Î≥¥Ï°∞ ÏßÄÌëú Î≥¥Ï†ï (RSI & MACD)
        # RSI < 30 (Í≥ºÎß§ÎèÑ) -> Í∞ïÏ†ú Îß§Ïàò Ïã†Ìò∏ (ÎπÑÏ§ë +0.3)
        if r_val < 30:
            target_ratio = max(target_ratio, 0.3)
            target_ratio += 0.2
            
        # RSI > 70 (Í≥ºÎß§Ïàò) -> ÎπÑÏ§ë Ï∂ïÏÜå
        if r_val > 70:
            target_ratio *= 0.5
            
        # MACD Í≥®Îì†ÌÅ¨Î°úÏä§ ÏÉÅÌÉú (MACD > Signal) -> ÎπÑÏ§ë ÌôïÎåÄ
        if m_val > m_sig:
            target_ratio *= 1.2
            # ÏµúÏÜå 10%Îäî Í∞ÄÏ†∏Í∞ÄÎèÑÎ°ù
            if target_ratio < 0.1: target_ratio = 0.1
        else:
            # Îç∞ÎìúÌÅ¨Î°úÏä§ -> ÎπÑÏ§ë Ï∂ïÏÜå
            target_ratio *= 0.8
            
        # 3. Î¶¨Ïä§ÌÅ¨ Í¥ÄÎ¶¨ (Ï¥àÍ≥†Î≥ÄÎèôÏÑ±Îßå ÌöåÌîº)
        if vol > vol_thresh:
            target_ratio *= 0.5
            
        # ÏµúÏ¢Ö ÎπÑÏ§ë ÌÅ¥Î¶¨Ìïë 0~1
        target_ratio = min(max(target_ratio, 0.0), 1.0)
            
        # 4. Î¶¨Î∞∏Îü∞Ïã±
        target_btc_val = portfolio_val * target_ratio
        current_btc_val = btc * price
        diff = target_btc_val - current_btc_val
        
        # Í±∞Îûò Ïã§Ìñâ
        if diff > 0: # Îß§Ïàò
            # 10Îã¨Îü¨ Ïù¥ÏÉÅÏùº ÎïåÎßå Í±∞Îûò
            if diff > 10:
                amount_to_buy_usd = diff
                if amount_to_buy_usd > cash: amount_to_buy_usd = cash
                if amount_to_buy_usd > 0:
                    btc_bought = (amount_to_buy_usd * (1 - tx_fee)) / price
                    btc += btc_bought
                    cash -= amount_to_buy_usd
                    trade_log.append({'date': dates[i], 'action': 'BUY', 'price': price, 'value': amount_to_buy_usd})
        elif diff < 0: # Îß§ÎèÑ
            if -diff > 10:
                amount_to_sell_usd = -diff
                amount_to_sell_btc = amount_to_sell_usd / price
                if amount_to_sell_btc > btc: amount_to_sell_btc = btc
                if amount_to_sell_btc > 0:
                    cash_gained = amount_to_sell_btc * price * (1 - tx_fee)
                    cash += cash_gained
                    btc -= amount_to_sell_btc
                    trade_log.append({'date': dates[i], 'action': 'SELL', 'price': price, 'value': cash_gained})
            
        history.append(cash + btc * price)

    total_return = (history[-1] - initial_capital) / initial_capital * 100
    
    return {
        'initial_capital': initial_capital,
        'final_value': history[-1],
        'total_return': total_return,
        'portfolio_values': history,
        'num_trades': len(trade_log),
        'total_fees_paid': 0 # Simplified
    }

# ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
my_result = simulate_custom_strategy(
    probs=my_prob,
    prices=test_prices,
    dates=test_dates,
    rsi=test_rsi,
    macd=test_macd,
    macd_sig=test_macd_sig,
    volatilities=test_volatility,
    vol_thresh=volatility_threshold
)

print("="*70)
print("üöÄ ÎÇòÏùò Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ Í≤∞Í≥º (Aggressive Hybrid)")
print("="*70)
print(f"Ï¥àÍ∏∞ ÏûêÎ≥∏: ${my_result['initial_capital']:,.2f}")
print(f"ÏµúÏ¢Ö ÏûêÎ≥∏: ${my_result['final_value']:,.2f}")
print(f"ÏàòÏùµÎ•†: {my_result['total_return']:.2f}%")
print(f"Í±∞Îûò ÌöüÏàò: {my_result['num_trades']}Ìöå")
print("="*70)
````

## ‚úÖ Í≥ºÏ†ú Ï≤¥ÌÅ¨Î¶¨Ïä§Ìä∏

- [‚úÖ] ÌïôÏÉù Ï†ïÎ≥¥ (Ïù¥Î¶Ñ, ÌïôÎ≤à) ÏûëÏÑ±
- [‚úÖ] MyTradingModel ÌÅ¥ÎûòÏä§ Íµ¨ÌòÑ ÏôÑÎ£å

- [‚úÖ] Ìä∏Î†àÏù¥Îî© Ï†ÑÎûµ ÏÑ§Í≥Ñ Î∞è ÏãúÎÆ¨Î†àÏù¥ÏÖò
